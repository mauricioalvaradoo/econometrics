## 1. Definiciones básicas

- Ensayos o experimentos

Cualquier acción o procesos cuyo resultado se encuentra sujeto a incertidumbre. En términos formales, cualquier recopilación (subconjunto) de resultados contenidos en un espacio muestral. Un ensayo puede ser tan simple como lanzar una moneda o sacar una carta de una baraja, o tan complejo como observar si una persona diagnosticada con cáncer de mama sigue viva cinco años después del diagnóstico. A lo largo de esta sección se les denotará con letra mayúscula: $A$, $B$, etc.  

- Espacio muestral

Es el conjunto de todos los posibles resultados elementales de un ensayo. Generalmente, se representa por $S$. Si un ensayo consiste en lanzar una moneda al aire, entonces un espacio muestral es $S=\{cara, sello\}$ o $S=\{c,s\}$ porque esas dos alternativas representan todos los resultados posibles de un experimento. Estos resultados tambien son conocidos como puntos de muestral. Si el experimento consta de múltiples ensayos, todas las combinaciones posibles de resultados de los ensayos deben especificarse como parte del espacio muestral. A modo de ejemplo, si se lanza una moneda al aire dos veces el especio muestral estaría definido por $S=\{(c,c),(c,s),(s,c),(s,s)\}$. 
 


- Eventos 

Es cualquier recopilación (subconjunto) de resultados contenidos en el espacio muestral $S$. Un evento, es representado por $E$ o cualquier letra mayúscula que no sea $S$, es la especificación del resultado de un juicio y puede consistir en un único resultado o en un conjunto de resultados. Por ejemplo, el evento de "cara al lanzar una moneda" podría especificarse como $E = \{cara\}$, mientras que el evento de "número impar al lanzar un dado" podría especificarse como $E = \{1, 3, 5\}$. A continuación los tipos de eventos: 

1. Evento seguro: siempre ocurre, pues tienen todos los resultados que se pueden presentar.

2. Evento imposible: es un conjunto vacío. No posee elementos, ya que nunca ocurre.

3. Evento mutuamente excluyente: si existe dos eventos $A$ y $B$ se dice que son mutuamente excluyentes si carecen de elementos en común. Es decir, $A$ y $B$ nunca ocurre de forma simultanea. Formalmente $A \cap B = \emptyset $

## 2. Teoría de conjuntos

Un evento como tal es un conjunto. Las relaciones y resultados de la teória elemental de conjuntos pueden ser utilizados para ejemplificar los eventos. A continuación se usaran las siguientes operaciones para crear y definir algunos eventos dados. 

- Unión

La unión de varios eventos simples crea un evento compuesto que ocurre si ocurre uno o más de los eventos. La unión de $E$ y $F$ se escribe $E \cup F$ y significa “o $E$ o $F$ o ambos $E$ y $F$”. Para tomar un ejemplo, suponga que el evento es lanzar un dado de seis caras y que $E = \{1, 3\}$ y $F = \{1, 2\}$. El evento $E \cup F$ se satisface con un resultado de 1, 2 o 3; también podemos decir que $E \cup F = \{1, 2, 3\}$.

La unión de E y F es el área sombreada en el diagrama de Venn a continuación. Tenga en cuenta que esta figura representa dos círculos completos que se superponen parcialmente; el significado de este diagrama es que cualquier punto en el área sombreada (cualquier punto en $E$, $F$, o ambos $E$ y $F$) satisface la condición $E \cup F$.

```python
sns.set_style("whitegrid")
v.true_false(subs=2, size=5, fill_color='lightblue', title_a='$A \cup B$ ',
           font_size=25, set_a='P', set_b='Q')
```

- Intersección

La intersección de dos o más eventos simples crea un evento compuesto que ocurre solo si ocurren todos los eventos simples. La intersección de $E$ y $F$ se escribe $E \cap F$ y significa "tanto $E$ como $F$".  Para continuar con el ejemplo, si el evento es lanzar un dado de seis caras, y $E = \{1, 3\}$ y $F = \{1, 2\}$, el evento $E \cap ​​F$ se satisface solo con el resultado de 1 porque 1 es miembro de ambos conjuntos, entonces $E \cap F = \{1\}$.

La intersección de $E$ y $F$ es el área sombreada en el diagrama de Venn en la siguiente figura. Se debe tener en cuenta que solo los puntos que pertenecen tanto a $E$ como a $F$ satisfacen la condición.

```python
v.and_nand(subs=2, size=5, fill_color='lightblue', 
         font_size=25, title_a='$A \cap B$', set_a='A', set_b='B')
```

- Complemento

El complemento de un evento significa todo en el espacio muestral que no es evento estudiado. El complemento del evento E se escribe como $~E$, $E^c$ o $Ē$, y se lee como “no $E$” o “complemento de $E$”. Por ejemplo, si $E = \{numeros>0\}$, su complemento seria $E^c=\{numero \leqslant 0 \}$.

Continuando, si el evento es lanzar un dado de seis caras y $E = \{1, 3\}$, $~E = \{2, 4, 5, 6\}$. El complemento de $F$ es el área sombreada en el diagrama de Venn de la siguiente figura.

```python
v.not_not(title_a = '$A^c$', title_b='$B^c$', set_a='A', set_b='B', fill_color='lightblue')
```

- Mutuamente excluyente

Si los eventos no pueden ocurrir juntos, son mutuamente excluyentes. Para decirlo de otra manera, si dos conjuntos no tienen eventos en común, son mutuamente excluyentes. Por ejemplo, el evento $A =$ (el salario es mayor a 100) y el evento $B =$ (el salario es menor o igual a 100) son mutuamente excluyentes, al igual que los conjuntos $A =$ (enteros pares) y $B =$ (enteros impares). 

Los conjuntos mutuamente excluyentes E y F se presentan en el diagrama de Venn de la figura 2-6; tenga en cuenta que no tienen puntos en común.

```python
v.mut_exclusive(size=10, fill_color='lightblue', bg_color='white', font_size=25, 
        title='$P \cap Q=∅$ ', set_a='P', set_b='Q')
```

- Independencia

Si dos ensayos son independientes, el resultado de un ensayo no influye en el resultado de otro. Para decirlo de otra manera, si los ensayos son independientes, conocer el resultado de un ensayo no te da información sobre el resultado del otro. El ejemplo clásico de independencia es lanzar una moneda ordinaria; si lanza la moneda dos veces, el resultado de la primera prueba no influye en el resultado de la segunda prueba.
