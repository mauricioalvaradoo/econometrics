# **I. Distribuciones**

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import scipy.stats as stats
import seaborn as sns

np.random.seed(10)
sns.set_style("whitegrid")
```

## **1. Distribuciones de probabilidad para V.A Discretas**

### **1.1. Distribución Bernoulli**

Esta es la distribución que se presenta cuando contamos el número $X$ de veces que ocurre un determinado evento $A$ sobre un experimento binario (conocido también como de Bernoulli).  Es decir, de un experimento en el cual existe una probabilidad $p$ de que el evento $A$ ocurra y una probabilidad $1-p$ de que este evento no ocurra.

Una v.a. discreta con rango $R_X = \{0, 1\}$ se dice que tiene distribución bernoulli de parámetro $p$, y se le denota por $X ∼ Ber(p)$, si su función de probabilidad viene dada por:

$$ p(X=x)=p^x(1-p)^{1-x};x=0,1$$

Donde: $ E(X)=p; V(X) = p (1-p) $.

```python
np.random.seed(10)

p = 0.3  
samples = np.random.binomial(1, p, size=1000)

plt.hist(samples, bins=[0, 1, 2], align='left', rwidth=0.9,
         alpha=0.75, color='blue', edgecolor='black',
         )

plt.xticks([0, 1])
plt.xlabel('Resultado (0 o 1)')
plt.ylabel('Frecuencia')
plt.title('Distribución Bernoulli (p=0.3)')

plt.show()
```

### **1.2. Distribución Binomial**

Esta es la distribución que se presenta cuando contamos el número $X$ de veces que ocurre un determinado evento $A$ sobre un total fijo de $n$ repeticiones u observaciones independientes de un experimento experimento binario (conocido también como de Bernoulli); es decir, de un experimento en el cual existe una probabilidad $p$ de que el evento $A$ ocurra y una probabilidad $1-p$ de que este evento no ocurra.



Una v.a. discreta con rango $R_X = \{0, 1, 2, . . . , n\}$ se dice que tiene distribución binomial de parámetros $n$ y $p$, y se le denota por $X ∼ B(n, p)$, si su función de probabilidad viene dada por: 

$$ P(X = x) = \binom{n}{x}p^x(1-p)^{n-x};x=0,1,2,...n $$

$$ \binom{n}{x} = \frac{n!}{x!(n-x)!} $$

Donde: $E(X)=np; V(X)=np(1-p)$.

```python
size = 1000
sample = np.random.binomial(20, 0.5, size)
bin = np.arange(0,20,1)

plt.hist(sample, bins=bin, alpha=0.75, align='left', edgecolor='black', color='blue', rwidth=0.9) 

plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Distribución Binomial $(n=20, p=0.5)$')

plt.show()
```

```python
size = 1000
sns.kdeplot(np.random.binomial(15, 0.5, size))
sns.kdeplot(np.random.binomial(20, 0.5, size))
sns.kdeplot(np.random.binomial(25, 0.5, size))


plt.ylabel('Densidad')
plt.title('Distribución Binomial (pmf)')
plt.legend(["$n = 15, p = 0.5$", 
            "$n = 20, p = 0.5$", 
            "$n = 25, p = 0.5$"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.binomial(15, 0.5, size))
sns.ecdfplot(np.random.binomial(20, 0.5, size))
sns.ecdfplot(np.random.binomial(25, 0.5, size))

plt.ylabel('Proporción')
plt.title('Distribución Binomial (cmf)')
plt.legend(["$n = 15, p = 0.5$", 
            "$n = 20, p = 0.5$", 
            "$n = 25, p = 0.5$"])
plt.show()
```

### **1.3. Distribución Geométrica**

La distribución geométrica aparece como resultado de contar cuántas veces se debe repetir de manera independiente un experimento hasta lograr que ocurra un determinado suceso $A$ por primera vez. Por ello, es útil cuando tenemos interés en el número de intentos necesarios hasta que ocurra el primer éxito.


- Considere un experimento aleatorio de cuyo espacio muestral $S$, se tienen un evento $A$ que puede ocurrir con probabilidad $p$. De igual forma, la probabilidad de no ocurrir $A$ es $(1-p)$. 

- Si se repite este experimento de manera independiente y se define la v.a discreta como $X =$ número de experimentos hasta que ocurra evento $A$, entonces la función de probabilidad de $X$ viene dada por la distribución geométrica. 

Una v.a. discreta con rango $R_X = {1, 2, . . . , }$ se dice que tiene distribución geométrica de parámetro $p$, y se le denota por $X ∼ Geo(p)$, si su función de probabilidad viene dada por:

$$ P(X=x)=p(1-p)^{x-1}; x=1,2,...$$

Donde: $E(X) = \frac{1}{p}; V(X) = \frac{(1-p)}{p^2}$



```python
size = 1000

sample = np.random.geometric(0.5, size)
bin = np.arange(0,20,1)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.title("Distribución Geométrica $(p = 0.5)$") 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.geometric(0.2, size))
sns.ecdfplot(np.random.geometric(0.5, size))
sns.ecdfplot(np.random.geometric(0.8, size))


plt.ylabel('Proporción')
plt.title('Distribución Geométrica (cdf)')
plt.legend(["$p = 0.2$", 
            "$p = 0.5$", 
            "$p = 0.8$"])
plt.show()
```

### **1.4. Distribución Binomial Negativa**

Es una generalización de la distribución geométrica, que surge cuando se repite el  experimento aleatorio en ella hasta que ocurra el evento $A$ por $r$-ésima vez, siendo $r$ un entero positivo de valor fijo. Por ello es útil cuando tenemos interés en el número de intentos necesarios hasta que ocurra el r-ésimo éxito

- Considere un experimento aleatorio de cuyo espacio muestral $S$, se tiene un evento $A$ que puede ocurrir con probabilidad $p$ y  de no ocurrir con probabilidad $1-p$. 
- Si se repite este experimento de manera independiente y se define la v.a. discreta: $X =$ número de experimentos hasta que ocurra $r$-veces el evento $A$, entonces la función de probabilidad de $X$ viene dada por la distribución binomial negativa. 

Una v.a. discreta con rango $R_X = \{r, r+1, r+2, . . . , \}$ se dice que tiene distribución binomial negativa de parámetros $r$ y $p$, y se le denota por $X ∼ BN(r, p)$, si su función de probabilidad viene dada por: 

$$ P(X=x)= \binom {x-1}{r-1} p^r (1-p)^{x-r}; x=r,r+1,r+2, ...$$

Donde: $E(X)=\frac {r}{p}; V(X) = \frac {r(1-p)}{p^2}$

Asimismo, si $r=1$, entonces $X ∼ Geo(p)$. 


```python
size = 1000

sample = np.random.negative_binomial(20, 0.5, size)
bin = np.arange(0,20,1)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.title("Distribución Binomial Negativa $(r=20, p=0.5)$") 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
size = 1000
sns.kdeplot(np.random.negative_binomial(15, 0.5, size))
sns.kdeplot(np.random.negative_binomial(20, 0.5, size))
sns.kdeplot(np.random.negative_binomial(25, 0.5, size))


plt.ylabel('Densidad')
plt.title('Distribución Binomial Negativa (pmf)')
plt.legend(["$r = 15, p = 0.5$", 
            "$r = 20, p = 0.5$", 
            "$r = 25, p = 0.5$"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.negative_binomial(15, 0.5, size))
sns.ecdfplot(np.random.negative_binomial(20, 0.5, size))
sns.ecdfplot(np.random.negative_binomial(25, 0.5, size))


plt.ylabel('Proporción')
plt.title('Distribución Binomial Negativa (cdf)')
plt.legend(["$p = 0.2 $ y $r = 15$", 
            "$p = 0.5 $ y $r = 20$", 
            "$p = 0.8 $ y $r = 25$"])
plt.show()
```

### **1.5. Distribución Hipergeométrica**

Es útil cuando tenemos interés en muestreo aleatorio sin reemplazo. Supongamos lo siguiente, sea:

- $N$ : numero de elementos de una población
- $M$ : elementos de N que pertenecen a una categoría A. 
- $n$ : elementos sin reemplazo de N. 
- $x$ : elementos de n que pertenecen a una categoría A. 

La distribución geométrica mide la probabilidad de obtener x elementos de la categoría $A$ en una muestra sin reemplazo de $n$ elementos de $N$. Es decir, se definel la v.a. $X =$ número de elementos de la categoría $A$ en
una muestra sin reemplazo de $n$ elementos de un total de $N$ elementos en la población. 

Una v.a. discreta con rango $R_X = \{1, . . . , n\}$ se dice que tiene distribución hipergeométrica de parámetros $N$,$M$, y $n$, y se le denota por $X ∼ H(N, M, n)$, si su función de probabilidad viene dada por: 

$$ P(X=x)=\frac {\binom {M}{x} \binom {N-M}{n-x}}{\binom {N}{n}}; x = 0,1,..., n$$

Donde: $E(X)=n \frac {M}{N}; V(X) = n \frac{M}{N} \frac {N-M}{N} \frac {N-n}{N-1}$

```python
size = 1000

sample = np.random.hypergeometric(20,19,10, size)
bin = np.arange(0,10,1)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.title("Distribución Hipergeométrica $(N=20, M=19, n=10)$") 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
size = 1000

sns.kdeplot(np.random.hypergeometric(20,12,10, size))
sns.kdeplot(np.random.hypergeometric(20,15,20, size))
sns.kdeplot(np.random.hypergeometric(20,18,30, size))

plt.ylabel('Densidad')
plt.title('Distribución Hipergeométrica (pmf)')
plt.legend([r"$\lambda = 5$", 
            r"$\lambda = 10$", 
            r"$\lambda = 15$"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.hypergeometric(20,12,10, size))
sns.ecdfplot(np.random.hypergeometric(20,15,20, size))
sns.ecdfplot(np.random.hypergeometric(20,18,30, size))


plt.ylabel('Proporción')
plt.title('Distribución Poisson (cdf)')
plt.legend([r"$\lambda = 5$", 
            r"$\lambda = 10$", 
            r"$\lambda = 15$"])
plt.show()
```

### **1.6. Distribución Poisson**

Es útil cuando estamos interesados en el número de ocurrencias de un evento en un determinado intervalo o periodo de tiempo fijo. Luego, se define la v.a. $X =$ número de ocurrencias del evento $A$ en un intervalo de tiempo.


Una v.a. discreta con rango $R_X = \{1, . . . , \}$ se dice que tiene distribución Poisson de parámetro $λ$, y se le denota por $X ∼ P(λ)$, si su función de probabilidad viene dada por: 

$$ P(X=x)= \frac {e^{-λ} λ^x}{x!}; x= 0,1,...$$

Donde: $E(X)= λ; V(X)=λ$ 
Asimismo, $λ$ es el número promedio (medio) de ocurrrencias del evento A en un intervalo de tiempo determinado. 

```python
size = 1000

sample = np.random.poisson(10, size)
bin = np.arange(0,20,1)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.title("Distribución Poisson $ (\lambda = 10)$") 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
size = 1000

sns.kdeplot(np.random.poisson(5, size))
sns.kdeplot(np.random.poisson(10, size))
sns.kdeplot(np.random.poisson(15, size))

plt.ylabel('Densidad')
plt.title('Distribución Poisson (pmf)')
plt.legend([r"$\lambda = 5$", 
            r"$\lambda = 10$", 
            r"$\lambda = 15$"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.poisson(5, size))
sns.ecdfplot(np.random.poisson(10, size))
sns.ecdfplot(np.random.poisson(15, size))


plt.ylabel('Proporción')
plt.title('Distribución Poisson (cdf)')
plt.legend([r"$\lambda = 5$", 
            r"$\lambda = 10$", 
            r"$\lambda = 15$"])
plt.show()
```

## **2. Distribuciones de probabilidad para V.A Continuas**

### **2.1. Distribución Uniforme**

Es útil cuando no hay razón para asignar probabilidades diferentes a un “conjunto de posibles valores de la v.a. en un determinado intervalo”. Es decir la probabilidad de cualesquiera dos intervalos de la misma longitud es la misma.

Una v.a. continua con rango $R_X = [a, b]$ se dice que tiene distribución uniforme de parámetros $a$ y $b$, y se le denota por $X ∼ Unif(a, b)$, si su función de densidad viene dada por:

$$ f_X(x) = \frac {1}{b-a}; $$

Donde: $E(X) = \frac {a+b}{2}; V(X) = \frac {(b-a)^2}{12}; F_X(x)=\frac {x-a}{b-a}$

```python
size = 1000

sample = np.random.uniform(0, 1, size)
bin = np.arange(-1,2,0.05)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.title("Distribución Uniforme $(a=0, b=1)$") 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
size = 1000

sns.kdeplot(np.random.uniform(0, 1, size))
sns.kdeplot(np.random.uniform(0, 2, size))
sns.kdeplot(np.random.uniform(0, 3, size))

plt.ylabel('Densidad')
plt.title('Distribución Uniforme (pmf)')
plt.legend(["a = 0, b = 1", 
            "a = 0, b = 2", 
            "a = 0, b = 3"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.uniform(0, 1, size))
sns.ecdfplot(np.random.uniform(0, 2, size))
sns.ecdfplot(np.random.uniform(0, 3, size))

plt.ylabel('Proporción')
plt.title('Distribución Uniforme (cdf)')
plt.legend(["a = 0, b = 1", 
            "a = 0, b = 2", 
            "a = 0, b = 3"])
plt.show()
```

### **2.2. Distribución Normal**

Una v.a. continua con rango $R_X = [−∞, ∞]$ se dice que tiene distribución normal de parámetros $µ$ (media) y $σ^2$ (varianza), y se le denota por $X ∼ N(µ, σ^2)$, si su función de densidad viene dada por: 

$$ f_X(x) = \frac {1}{\sqrt{2\pi}σ^2}exp -\frac {1}{2}(x-µ)^2; $$

Donde: $E(X)= µ; V(X)=σ^2$

Asimismo, se debe tener en cuenta lo siguiente: 
- La función de densidad de una v.a. Normal tiene forma de campana, es simétrica alrededor de la media $µ$. Además la $µ = media = mediana = moda$.
- Es una v.a. definida en todos los reales.
- La varianza $σ^2$ controla la dispersión de los datos alrededor de la media $µ$. Si $µ$ se mantiene fija y $σ^2$ crece, la distribución se “aplana”; en cambio si $σ^2$ disminuye, la distribución se “angosta”.
- $σ$ se llama desviación estándar.

```python
size = 1000

sample = np.random.normal(0, 1, size)
bin = np.arange(-5,5,0.1)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title("Distribution Normal ($µ = 0, σ =1$) ") 
plt.show()
```

```python
size = 1000

sns.kdeplot(np.random.normal(0, 1, size))
sns.kdeplot(np.random.normal(0, 2, size))
sns.kdeplot(np.random.normal(0, 3, size))

plt.ylabel('Densidad')
plt.title('Distribución Normal (pmf)')
plt.legend([r"$\mu = 0, \sigma = 1$", 
            r"$\mu = 0, \sigma = 2$", 
            r"$\mu = 0, \sigma = 3$"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.normal(0, 1, size))
sns.ecdfplot(np.random.normal(0, 2, size))
sns.ecdfplot(np.random.normal(0, 3, size))

plt.ylabel('Proporción')
plt.title('Distribución Normal (cdf)')
plt.legend([r"$\mu = 0, \sigma = 1$", 
            r"$\mu = 0, \sigma = 2$", 
            r"$\mu = 0, \sigma = 3$"])
plt.show()
```

### **2.3. Distribución Exponencial**

Es útil cuando la v.a. es continua y los valores observados que la v.a. asume son valores no negativos. Por ejemplo, para describir el tiempo que se necesita para completar una tarea, tiempo de llegada de clientes a un banco, vida útil de un bien, tiempo continuos de falla, etc. 

Una v.a.a continua con rango $R_X = [−∞, ∞]$ se dice que tiene distribución exponencial de parámetros $\lambda$, y se le denota por $X ∼ Exp(λ)$, si su función de densidad viene dad por: 

$$  f_X(x) = \lambda e^{-\lambda x} $$

Donde:  $E(X) = \frac {1}{\lambda}$, $V(X) = \frac {1}{\lambda^2}$ y $F_X(x) = 1-e^{-\lambda x}$

Asimismo, se debe tener en cuenta que el parámetro $\lambda$ es llamado tasa de ocurrencia del evento por unidad de tiempo. La función de densidad tiene la forma de una función exponencial, por eso el nombre de distribución exponencial. 

```python
size = 1000

sample = np.random.exponential(1, size)
bin = np.arange(0,10,0.1)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title("Distribution Exponencial ($µ = 0, σ =1$) ") 
plt.show()
```

```python
size = 1000

sns.kdeplot(np.random.exponential(1, size))
sns.kdeplot(np.random.exponential(2, size))
sns.kdeplot(np.random.exponential(3, size))

plt.ylabel('Densidad')
plt.title('Distribución Exponencial (pmf)')
plt.legend([r"$\beta = 1$", 
            r"$\beta = 2$", 
            r"$\beta = 3$"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.exponential(1, size))
sns.ecdfplot(np.random.exponential(2, size))
sns.ecdfplot(np.random.exponential(3, size))

plt.ylabel('Proporción')
plt.title('Distribución Exponencial (cdf)')
plt.legend([r"$\beta = 1$", 
            r"$\beta = 2$", 
            r"$\beta = 3$"])
plt.show()
```

### **2.4. Distribución Gamma**

Es una generalización de la distribución exponencial. Una v.a. continua con rango $R_X = [−∞, ∞]$ se dice que tienen una distribución gamma de parámetros $α$ y $β$, y se le denota por $X ∼ gamma(α, β)$, si su función de densidad viene dada por:
$$X ∼ G(α, β) → \frac{β^α}{\Gamma(α)} x^{a-1}e^{-βx}$$ 

Donde: $E(X)=\frac {α}{β}; V(X)=\frac {\alpha}{\beta^2}$

$\Gamma (.)$ representa la función gamma, y es definida por: $\Gamma(k)=(k-1)\times (k-2)\times...2\times 1$. En particular, $Exp(\beta)= G(1,\beta)$. 

**Observacion:** Tanto la distribución exponencial como la gamma se usan para modelar datos con distribuciones asimétricas, como ingresos, tiempo de vida, etc.  

```python
size = 1000

sample = np.random.gamma(1, 2, size)
bin = np.arange(0,9,0.1)

plt.hist(sample, bins=bin, alpha=0.75, color='blue', edgecolor='black', rwidth=0.9) 
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title("Distribution Gamma $(α, = 0, β =1)$ ") 
plt.show()
```

```python
size = 1000

sns.kdeplot(np.random.gamma(1,2, size))
sns.kdeplot(np.random.gamma(2,2, size))
sns.kdeplot(np.random.gamma(5,1, size))

plt.ylabel('Densidad')
plt.title('Distribución Gamma (pmf)')
plt.legend([r"$\alpha = 1, \beta = 2$", 
            r"$\alpha = 2, \beta = 2$", 
            r"$\alpha = 5 ,\beta = 1$"])
plt.show()
```

```python
size = 1000

sns.ecdfplot(np.random.gamma(1,2, size))
sns.ecdfplot(np.random.gamma(2,2, size))
sns.ecdfplot(np.random.gamma(5,1, size))

plt.ylabel('Proporción')
plt.title('Distribución Exponencial (cdf)')
plt.legend([r"$\alpha = 1, \beta = 2$", 
            r"$\alpha = 2, \beta = 2$", 
            r"$\alpha = 5 ,\beta = 1$"])
plt.show()
```
